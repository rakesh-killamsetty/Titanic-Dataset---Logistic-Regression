```python
# 1.seperate the dependent and independent variables
#2.split the dataset

# 3.train the model by using fit method
#4.predict
#5.Accuracy Metrics
```

```python
x=data1.drop('Survived',axis=1)
y=data1['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
```

Pascal case is used in Logistic Regression

```python
ypred=classifier.predict(X_test)
ypred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,ypred)
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,ypred)
```

```python
results=pd.DataFrame(columns=['Actual','Predicted'])
results['Actual']=Y_test
results['Predicted']=ypred
results=results.reset_index()
results['Id']=results.index
results
```

```python
sns.lineplot(x='Id',y='Actual',data=results.head(100))
sns.lineplot(x='Id',y='Predicted',data=results.head(100))
```
